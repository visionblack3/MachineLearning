```python
import opendatasets as od
import numpy as np
od.download("https://www.kaggle.com/competitions/house-prices-advanced-regression-techniques/data")
```

```python
import pandas as pd
data_df=pd.read_csv("house-prices-advanced-regression-techniques/train.csv")
data_df.isna().sum()
```

```python
num_data=data_df.select_dtypes(include='number')
cor_data=num_data.corr()
cor_data
```

```python
req_col=['TotRmsAbvGrd','GrLivArea', 'BsmtFullBath', 'BsmtHalfBath', 'FullBath',
       'HalfBath','TotalBsmtSF','1stFlrSF', '2ndFlrSF','LowQualFinSF','SalePrice','GarageArea','OverallQual']
data_req=data_df.loc[:,req_col]
data_req.loc[:,'TotalBath']=(data_req['BsmtFullBath'].fillna(0)+data_req['BsmtHalfBath'].fillna(0)+data_req['HalfBath'].fillna(0)+data_req['FullBath'].fillna(0))
data_req.loc[:,'Totalsqft']=(data_req['GrLivArea'].fillna(0)+data_req['TotalBsmtSF'].fillna(0)+data_req['1stFlrSF'].fillna(0)+data_req['2ndFlrSF'].fillna(0)+data_req['LowQualFinSF'].fillna(0))


train_df=data_req[['TotRmsAbvGrd','GarageArea','OverallQual','TotalBath','Totalsqft','SalePrice']]
train_df
```

```python
from sklearn.model_selection import train_test_split
train_set,test_set=train_test_split(train_df,test_size=0.2,random_state=99)
housing=train_set.drop("SalePrice",axis="columns")
housing_label=train_set['SalePrice']
print(housing)
print(housing_label)
```

```python
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
my_pipeline = Pipeline([
    ('imputer',SimpleImputer(strategy="median")),
    ('std_scaler',StandardScaler())
])
x_train = my_pipeline.fit_transform(housing)
y_train=housing_label
print(x_train)
print(y_train)
```

```python
data2_df=pd.read_csv("house-prices-advanced-regression-techniques/train.csv")
req_col=['TotRmsAbvGrd','GrLivArea', 'BsmtFullBath', 'BsmtHalfBath', 'FullBath',
       'HalfBath','TotalBsmtSF','1stFlrSF', '2ndFlrSF','LowQualFinSF','GarageArea','OverallQual']
data_req2=data2_df.loc[:,req_col]
data_req2.loc[:,'TotalBath']=(data_req2['BsmtFullBath'].fillna(0)+data_req2['BsmtHalfBath'].fillna(0)+data_req2['HalfBath'].fillna(0)+data_req2['FullBath'].fillna(0))
data_req2.loc[:,'Totalsqft']=(data_req2['GrLivArea'].fillna(0)+data_req2['TotalBsmtSF'].fillna(0)+data_req2['1stFlrSF'].fillna(0)+data_req2['2ndFlrSF'].fillna(0)+data_req2['LowQualFinSF'].fillna(0))
data_req2
```

```python
test_df_unproc = data_req2[['TotRmsAbvGrd','TotalBath','GarageArea','Totalsqft','OverallQual']]
x_test = my_pipeline.transform(test_df_unproc.values)
x_test
```

```python
print(x_train)
print(y_train)
print(x_test)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_train)
```

```python
def mse(target,predicted):
  return np.mean(np.square(target-predicted))
```

```python
train_mse = mse(y_train,y_pred)
train_rmse = np.sqrt(train_mse)

print(f"Training MSE: {train_mse:.2f}, Training RMSE: {train_rmse:.2f}")
```
